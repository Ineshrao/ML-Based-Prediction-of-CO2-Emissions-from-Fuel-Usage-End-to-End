# Feature Engineering

```python
from scipy import stats
data=df[['ENGINE SIZE', 'CYLINDERS','FUEL CONSUMPTION','TRANSMISSION','FUEL','HWY (L/100 km)','COMB (L/100 km)', 'COMB (mpg)','EMISSIONS']].copy()
data=data[(np.abs(stats.zscore(data[['ENGINE SIZE', 'CYLINDERS','FUEL CONSUMPTION','HWY (L/100 km)','COMB (L/100 km)', 'COMB (mpg)','EMISSIONS']])) < 3).all(axis=1)]
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
trans_col=make_column_transformer((OneHotEncoder(sparse=False),['TRANSMISSION','FUEL']),remainder='passthrough')
sc=StandardScaler()
```

# Training Model

1.   Linear Regression
2.   Polynomial Linear Regression
3.   Decision Tree Regressor
4.   Random Forest Regressor
5.   Support Vector Regressor


```python
from sklearn.model_selection import train_test_split

X = data.drop(['EMISSIONS'], axis=1)
y = log_data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
pipe_lr=make_pipeline(trans_col,sc,lr)
pipe_lr.fit(X_train,y_train)
y_pred_lr=pipe_lr.predict(X_test)
ytain_pred_lr=pipe_lr.predict(X_train)
```

```python
poly_reg = PolynomialFeatures(degree=2)

plr = LinearRegression()
pipe_poly_lr=make_pipeline(trans_col,poly_reg,sc,plr)
pipe_poly_lr.fit(X_train,y_train)
y_pred_poly_lr=pipe_poly_lr.predict(X_test)
ytain_pred_poly_lr=pipe_poly_lr.predict(X_train)
```

```python
y_pred_poly_lr=pipe_poly_lr.predict(X_test)
```

```python
dt = DecisionTreeRegressor(random_state=42)

pipe_dt=make_pipeline(trans_col,sc,dt)
pipe_dt.fit(X_train,y_train)
y_pred_dt=pipe_dt.predict(X_test)
ytain_pred_dt=pipe_dt.predict(X_train)
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
pipe_rf=make_pipeline(trans_col,sc,rf)
pipe_rf.fit(X_train,y_train)
y_pred_rf=pipe_rf.predict(X_test)
ytain_pred_rf=pipe_rf.predict(X_train)
```

```python
svr = SVR(kernel='linear')

pipe_svr=make_pipeline(trans_col,sc,svr)
pipe_svr.fit(X_train,y_train)
y_pred_svr=pipe_svr.predict(X_test)
ytain_pred_svr=pipe_svr.predict(X_train)
```

```python
print('R-squared TEST:')
print('Linear Regression : ', r2_score(y_test, y_pred_lr))
print('Polynomial Linear Regression :', r2_score(y_test,y_pred_poly_lr))
print('Decision Tree Regressor :', r2_score(y_test, y_pred_dt))
print('Random Forest Regressor:', r2_score(y_test, y_pred_rf))
print('Support Vector Regressor:', r2_score(y_test, y_pred_svr))



print('R-squared TRAIN:')
print('Linear Regression : ', r2_score(y_train, ytain_pred_lr))
print('Polynomial Linear Regression :', r2_score(y_train,ytain_pred_poly_lr))
print('Decision Tree Regressor :', r2_score(y_train, ytain_pred_dt))
print('Random Forest Regressor:', r2_score(y_train, ytain_pred_rf))
print('Support Vector Regressor:', r2_score(y_train, ytain_pred_svr))
```
